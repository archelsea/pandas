## Pclass 선택하기

```python
titanic['Pclass'].isin([2,3]) #SQL 'IN'과 비슷
```

```python
titanic[titanic['Pclass'].isin([2,3])]
```

```python
titanic['Pclass'] == 2 #pclass가 2일 때만 True가 나옴
```

```python
# OR 연산

titanic[(titanic['Pclass'] == 2) | (titanic['Pclass'] == 3)]   
```

```python
# AND 연산

titanic[(titanic['Pclass'] == 2) & (titanic['Age'] == 30)] 
```

# NULL값 처리하기

```python
titanic.info()
```

### null 확인하기




```python
# notna() = null이면 false, null이 아니면 true가 나옴
titanic['Age'].notna()
```

```python
age_notna =  titanic[titanic['Age'].notna()]
age_notna
# age columns 이 null아닌 것들로만 df만들어 줌
```

```python
# isna() = null이면 true, null이 아니면 false가 나옴
titanic['Age'].isna()
```

```python
# 각 컬럼을 다 더해줌
titanic.sum()
```

```python
# notna,isna는 series뿐만 아니라 df에서도 적용이 가능함

titanic.isna()
```

```python
titanic.isna().sum()
```

### null 채우기

```python
titanic_fillna_0 = titanic.fillna(0) # null대신 0으로 채워줌
titanic_fillna_0.info()
```

### null 제거하기

```python
titanic_drop = titanic.dropna()
titanic_drop
```

* 많은 산술연산자들이 null을 제외하고 계산한다는 것 꼭 기억해 두고 null값을 어떻게 처리할지 생각하기!
