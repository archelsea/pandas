```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# index_col=0: 맨 처음 컬럼을 인덱스로 사용하겠다는 뜻 
air = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2.csv', index_col=0)
air
```

```python
air.info()
# datetime 데이터 타입이 string으로 들어가 있음, datetime 타입의 데이터로 보고 싶을 때는?
```

```python
air_quality = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality
```

```python
air_quality.info()
```

#lineplot

```python
air_quality.plot()
```

```python
air_quality['station_paris'].plot() #series에서도 plot 메소드 사용 가능
```

# Scatter plot

```python
air_quality.plot.scatter(x='station_london',y='station_paris',alpha=0.5) #alpha : 투명도 option
```

# Box plot

```python
air_quality.plot.box()
```

# Subplot

```python
air_quality.plot.area()
```

```python
air_quality.plot.area(subplots=True)
```

# 그래프 저장하기

```python
fig,ax = plt.subplots(figsize=(12,4))
```

```python
air_quality.plot.area(ax=ax)
```

```python
#ylabel만들기
ax.set_ylabel('NO$_2$ concentration')
```
