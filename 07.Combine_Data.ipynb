```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2_long.csv')
air_quality_pm25 = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_pm25_long.csv')
```

# Concat

SELECT *    
FROM air_quality_pm25     
UNION     
SELECT *    
FROM air_quality_no2

```python
#axis=1(열) 옆으로 이어붙이기, axis=0(행) 아래로 이어붙이기

air_quality = pd.concat([air_quality_no2,air_quality_pm25],axis=0)
air_quality.head(2)
```

```python
#제대로 들어갔는지 확인
air_quality['parameter'].unique()
```

```python
#잘 붙엇는지 확인

print(air_quality_no2.shape)
print(air_quality_pm25.shape)
print(air_quality.shape)
```

```python
# Table의 출처를 알 수 있음 => 멀티 index

air_quality_2 = pd.concat([air_quality_no2,air_quality_pm25],axis=0,keys=['NO2','PM2.5'])
air_quality_2
```

```python
#위의 인덱스 활용
air_quality_2.loc['PM2.5']
```

```python
# 모든 인덱스를 컬럼으로 변형
air_quality_2.reset_index()
```

```python
#위의 테이블에서 level_0만 column으로 변형하고 싶다면(지정 인덱스만 컬럼으로 변형)
air_quality_2.reset_index(level=0)
```

# Merge

+ 기준이 되는 컬럼의 이름이 동일할 때

```python
stations_coord = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_stations.csv')
```

```python
stations_coord.head(2)
```

```python
air_quality_3 = pd.merge(air_quality_2,stations_coord, how='left',on='location')
air_quality_3
```

+ 기준이 되는 컬럼의 이름이 다를 때

```python
air_quality_parameter = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_parameters.csv')
```

```python
air_para = pd.merge(air_quality_3,air_quality_parameter,how='left',left_on='parameter',right_on='id')
air_para.head(2)
```
